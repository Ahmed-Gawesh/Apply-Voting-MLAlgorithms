# Selecting Importance_Features To y Process

# Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
final_df=scaler.fit_transform(X)
```

# Feature Selection 

```python
from sklearn.feature_selection import SelectKBest,chi2
best_features=SelectKBest(chi2,k=50)
X_selected=best_features.fit_transform(final_df,y)

#Get Feature Names
selected_features=X.columns[best_features.get_support()]
print(selected_features)

# put New Features in Frame 
X_new=X[selected_features]
print(X_new.shape)
```

# Spliting Data and build Models

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.2,shuffle=True,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,VotingClassifier
from sklearn.model_selection import StratifiedKFold,GridSearchCV,cross_val_score
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

models={
    "Logistic Regression":LogisticRegression(),
    "SVM":SVC(),
    "Decision Tree":DecisionTreeClassifier(),
    "naive bayes":GaussianNB()
}
ensemble_models={
    "Random Forest":RandomForestClassifier(),
    'Gradient Boosting':GradientBoostingClassifier(),
}

params={
    "Logistic Regression":{'C':[0.01,0.1,1,10]},
    "SVM":{ 'kernel':['rbf','linear'],'C':[0.01,0.1,1,10] },
    "Decision Tree":{'criterion':['gini','entropy'],'max_depth':[3,5,10]},
    "naive bayes":{},
    "Random Forest":{'n_estimators':[100,150,200]},
    'Gradient Boosting':{'n_estimators':[100,150,200]},
}
```

```python
#Model Evaluation
cv=StratifiedKFold(n_splits=5,random_state=0,shuffle=True)

best_models={}
# For Models 
for name, model in models.items():
    grid_search=GridSearchCV(model,params[name],cv=cv,scoring='accuracy')
    grid_search.fit(X_train,y_train)
    best_models[name]=grid_search.best_estimator_
    print(f"{name}: best parameters :{grid_search.best_params_} ,Best Score:{grid_search.best_score_} ")

# For Ensemble Models
for name, model in ensemble_models.items():
    grid_search=GridSearchCV(model,params[name],cv=cv,scoring="accuracy")
    grid_search.fit(X_train,y_train)
    best_models[name]=grid_search.best_estimator_
    print(f"{name}: best parameters :{grid_search.best_params_} ,Best Score:{grid_search.best_score_} ")

#--------------------------------------------------------
# Voting Method

estimators=[('rf',best_models['Random Forest']),('nb',best_models['naive bayes'])]
voting_clf=VotingClassifier(estimators=estimators, voting='soft')
voting_clf.fit(X_train,y_train)
best_models['Voting']=voting_clf

#---------------------------------------------------------
#Evaluate Best Model
for name,model in best_models.items():
    y_pred=model.predict(X_test)
    acc_score=accuracy_score(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    cm=confusion_matrix(y_test,y_pred)
    print(f"Name: {name}")
    print(f"Confusion Metrix: {cm}")
    print(f"Accuracy: {acc_score}")
    print(f"Report: {report}")
    
```
